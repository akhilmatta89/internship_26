# Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
# Intializing the chrome driver
driver = webdriver.Chrome('/Users/akhilmatta/chromedriver')
#Loading the youtube wikepedia URL
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
time.sleep(5)
# Loading table parent tag
table_prnt_tag_elem = driver.find_element_by_xpath('/html/body/div[3]/div[3]/div[5]/div[1]/table[4]')
# Loading tr child tag of table_prnt_tag_elem
child_tags = table_prnt_tag_elem.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[4]/tbody/tr')
video_name = []
rank = []
uploader_name = []
views_list=[]
upload_date_list = []
vid_name_count = 1
# looping through all the child tags
for i in child_tags:
    
    # Extracting video name from child tag
    try:
        video_name.append(i.find_element_by_xpath(f'//*[@id="mw-content-text"]/div[1]/table[4]/tbody/tr[{vid_name_count}]/td[1]/a').text)  
    except NoSuchElementException:
        pass
    rank.append(vid_name_count)
    # Extracting uploader name from child tag
    try:
        uploader_name.append(i.find_element_by_xpath(f'//*[@id="mw-content-text"]/div[1]/table[4]/tbody/tr[{vid_name_count}]/td[2]/a').text)
    except NoSuchElementException:
        try:
            uploader_name.append(i.find_element_by_xpath(f'//*[@id="mw-content-text"]/div[1]/table[4]/tbody/tr[{vid_name_count}]/td[2]').text)
        except NoSuchElementException:
            pass
    # Extracting views from child tag
    try:
        views_list.append(i.find_element_by_xpath(f'//*[@id="mw-content-text"]/div[1]/table[4]/tbody/tr[{vid_name_count}]/td[3]').text)
    except NoSuchElementException:
            pass
    # Extracting upload_date from child tag
    try:
        upload_date_list.append(i.find_element_by_xpath(f'//*[@id="mw-content-text"]/div[1]/table[4]/tbody/tr[{vid_name_count}]/td[4]').text)
    except NoSuchElementException:
            pass 
    # Incresing the count to extract items in next loop
    vid_name_count=vid_name_count+1
driver.close()
youtube_df = pd.DataFrame({})
youtube_df['Rank']=rank
youtube_df['Video Name'] = video_name
youtube_df['Artist']=uploader_name
youtube_df['Upload date']=upload_date_list
youtube_df['Views']=views_list
youtube_df
```

# Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Match title (I.e. 1
st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
# Intializing the chrome driver
driver = webdriver.Chrome('/Users/akhilmatta/chromedriver')
#Loading the https://www.bcci.tv/ URL
driver.get('https://www.bcci.tv/')
time.sleep(5)
# Getting fixture href link
fix_link = driver.find_element_by_xpath('/html/body/nav/div[1]/div[2]/ul[1]/li[2]/a').get_attribute('href')
driver.get(fix_link)
time.sleep(5)
# Getting parent tag
matches_body = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div/div/div/div[2]/div[3]/div')
# Getting all child tags related to parent tags
get_matches_tag=matches_body.find_elements_by_xpath('//*[@id="fixtures"]/div[3]/div/div')
Match_title = []
match_loc = []
match_series = []
match_date = []
match_time = []
matchnum=1
# Looping through all child tags
for i in get_matches_tag:
    # Getting match title
    Match_title.append(i.find_element_by_xpath(f'/html/body/div[2]/div[2]/div/div/div/div[2]/div[3]/div/div[{matchnum}]/div/div[4]/div/span[1]').text.replace(" -",""))
    # Extracting match location
    match_loc.append(i.find_element_by_xpath(f'/html/body/div[2]/div[2]/div/div/div/div[2]/div[3]/div/div[{matchnum}]/div/div[4]/div/span[2]').text)
    # Extracting series names
    match_series.append(i.find_element_by_xpath(f'//*[@id="fixtures"]/div[3]/div/div[{matchnum}]/div/div[1]/h5[2]/span').text)
    # Extracting match date
    match_date.append(i.find_element_by_xpath(f'//*[@id="fixtures"]/div[3]/div/div[{matchnum}]/div/div[1]/div/div[1]/h5').text)
    # Extracting match time
    match_time.append(i.find_element_by_xpath(f'//*[@id="fixtures"]/div[3]/div/div[{matchnum}]/div/div[1]/div/div[2]/h5').text)
    matchnum+=1
driver.close()
# Ceating a DataFrame
bcci_df = pd.DataFrame({})
bcci_df['Match title']=Match_title
bcci_df['Series'] = match_series
bcci_df['Place']=match_loc
bcci_df['Date']=match_date
bcci_df['Time']=match_time
bcci_df
```

D) Language used

```python
# Intializing the chrome driver
driver = webdriver.Chrome('/Users/akhilmatta/chromedriver')
#Loading the https://github.com/ URL
driver.get('https://github.com/')
time.sleep(5)
driver.get(driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a').get_attribute('href'))
```

```python
repo_tags = driver.find_elements_by_xpath('/html/body/div[4]/main/div[3]/div/div[2]/article')
repo_names = []
repo_desc = []
repo_contr = []
repo_lang = []
repo_rows = 1
for i in repo_tags:
    repo_names.append(i.find_element_by_xpath(f'//*[@id="js-pjax-container"]/div[3]/div/div[2]/article[{repo_rows}]/h1/a').text)
    repo_desc.append(i.find_element_by_xpath(f'//*[@id="js-pjax-container"]/div[3]/div/div[2]/article[{repo_rows}]/p').text)
    if i.find_elements_by_xpath(f'//*[@id="js-pjax-container"]/div[3]/div/div[2]/article[{repo_rows}]/div[2]/span[2]/a'):
        repo_contr.append(len(i.find_elements_by_xpath(f'//*[@id="js-pjax-container"]/div[3]/div/div[2]/article[{repo_rows}]/div[2]/span[2]/a')))
    else:
        repo_contr.append(len(i.find_elements_by_xpath(f'//*[@id="js-pjax-container"]/div[3]/div/div[2]/article[{repo_rows}]/div[2]/span[1]/a')))
    try:
        repo_lang.append(i.find_element_by_xpath(f'//*[@id="js-pjax-container"]/div[3]/div/div[2]/article[{repo_rows}]/div[2]/span[1]/span[2]').text)
    except NoSuchElementException:
        repo_lang.append('-')
    repo_rows+=1
driver.close()
github_df = pd.DataFrame({})
github_df['Repository Title']=repo_names
github_df['Repository Description'] = repo_desc
github_df['Contributors Count']=repo_contr
github_df['Language']=repo_lang
github_df
```
