```python
import time
elem = driver.find_element(By.TAG_NAME,'body')
start_time = time.time()
seconds = 500
while True:
    current_time = time.time()
    elem.send_keys(Keys.END) 
    time.sleep(3)
    elem.send_keys(Keys.PAGE_UP)
    elapsed_time = current_time - start_time
    if elapsed_time > seconds:
        print("Finished iterating in: " + str(int(elapsed_time))  + " seconds")
        break

    
```

```python
parnt_tag = driver.find_element(By.XPATH,'/html/body/div[2]/div/div[1]/div/div/main/div[1]/div[1]/div[6]')
child_tags = parnt_tag.find_elements(By.XPATH,"//div[@class = 'gsc_col-md-4 gsc_col-sm-6 gsc_col-xs-12  cardnew']")
```

```python
len(child_tags)
```

```python
count=1
block_hrefs = []
for i in child_tags:
    try:
        get_href = i.find_element(By.XPATH,f'/html/body/div[2]/div/div[1]/div/div/main/div[1]/div[1]/div[6]/div[{count}]/div/div[2]/div/div/a')
        block_hrefs.append(get_href.get_attribute('href'))
        count=count+1
    except NoSuchElementException as e:
        count=count+1
        pass
    
    
    
```

```python
car_names = []
car_make_year = []
car_variants = []
car_driven_kms = []
car_owners = []
car_transmission = []
car_price = []
location = []

for i in block_hrefs:
    driver.get(i)
    time.sleep(3)
    try:
        car_names.append(driver.find_element(By.XPATH,'/html/body/div[2]/div/div[1]/div/div/main/div/div[1]/div[2]/div[1]/div/div/div/div[1]/div[1]/h1').text)
    except NoSuchElementException as e:
        try:
            car_names.append(driver.find_element(By.XPATH,'/html/body/div[2]/div/div[1]/div/div/main/div[2]/div[1]/div/div[2]/div[2]/div/div/div/div[1]/div[1]/h1[2]').text)
        except NoSuchElementException as ee:
            car_names.append(np.nan)
    try:
        car_make_year.append(driver.find_element(By.XPATH,'/html/body/div[2]/div/div[1]/div/div/main/div/div[1]/div[1]/div[4]/div/div/div/ul/li[1]/div/div[2]').text)
    except NoSuchElementException as e:
        try:
            car_make_year.append(driver.find_element(By.XPATH,'/html/body/div[2]/div/div[1]/div/div/main/div[2]/div[1]/div/div[3]/div/div/div/div[1]/ul/li[1]/div/div').text)
        except NoSuchElementException as ee:
            car_make_year.append(np.nan)
    try:
        car_variants.append(driver.find_element(By.XPATH,'/html/body/div[2]/div/div[1]/div/div/main/div/div[1]/div[1]/div[4]/div/div/div/ul/li[3]/div/div[2]').text)
    except NoSuchElementException as e:
        try:
            car_variants.append(driver.find_element(By.XPATH,'/html/body/div[2]/div/div[1]/div/div/main/div[2]/div[1]/div/div[3]/div/div/div/div[1]/ul/li[5]/div/div').text)
        except NoSuchElementException as ee:
            car_variants.append(np.nan)
    try:
        car_driven_kms.append(driver.find_element(By.XPATH,'/html/body/div[2]/div/div[1]/div/div/main/div/div[1]/div[1]/div[4]/div/div/div/ul/li[4]/div/div[2]').text)
    except NoSuchElementException as e:
        try:
            car_driven_kms.append(driver.find_element(By.XPATH,'/html/body/div[2]/div/div[1]/div/div/main/div[2]/div[1]/div/div[3]/div/div/div/div[1]/ul/li[3]/div/div').text)
        except NoSuchElementException as ee:
            car_driven_kms.append(np.nan)    
    try:
        car_owners.append(driver.find_element(By.XPATH,'/html/body/div[2]/div/div[1]/div/div/main/div/div[1]/div[1]/div[4]/div/div/div/ul/li[6]/div/div[2]').text)
    except NoSuchElementException as e:
        try:
            car_owners.append(driver.find_element(By.XPATH,'/html/body/div[2]/div/div[1]/div/div/main/div[2]/div[1]/div/div[3]/div/div/div/div[1]/ul/li[4]/div/div').text)
        except NoSuchElementException as ee:
            car_owners.append(np.nan)  
    try:
        car_transmission.append(driver.find_element(By.XPATH,'/html/body/div[2]/div/div[1]/div/div/main/div/div[1]/div[1]/div[4]/div/div/div/ul/li[8]/div/div[2]').text)
    except NoSuchElementException as e:
        try:
            car_transmission.append(driver.find_element(By.XPATH,'/html/body/div[2]/div/div[1]/div/div/main/div[2]/div[1]/div/div[3]/div/div/div/div[1]/ul/li[6]/div/div').text)
        except NoSuchElementException as ee:
            car_transmission.append(np.nan)
    try:
        car_price.append(driver.find_element(By.XPATH,'/html/body/div[2]/div/div[1]/div/div/main/div/div[1]/div[2]/div[1]/div/div/div/div[1]/div[2]/div[1]/div[1]/span[2]').text)
    except NoSuchElementException as e:
        try:
            car_price.append(driver.find_element(By.XPATH,'/html/body/div[2]/div/div[1]/div/div/main/div[2]/div[1]/div/div[2]/div[2]/div/div/div/div[1]/span/span').text)
        except NoSuchElementException as ee:
            car_price.append(np.nan)
    location.append('Hyderabad')
```

```python
cardekho_df = pd.DataFrame({'Car Brand':car_names,'Make year': car_make_year,'Fuel Type': car_variants,'No of Owners':car_owners,'Transmission Type': car_transmission, 'Driven KMS': car_driven_kms, 'Location':location,'Car Price':car_price})


```
