# Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
url = 'https://www.imdb.com/search/title/?count=100&groups=top_100&sort=user_rating'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
movie_name = []
movie_year = []
rating = []
movie_data = soup.find_all('div', class_ = "lister-item mode-advanced")
for each_movie in movie_data:
    moviename=each_movie.h3.a.text
    movieyear = each_movie.h3.find('span', class_='lister-item-year text-muted unbold').text.replace('(','').replace(')','')
    movierating = each_movie.find('div', class_ = 'inline-block ratings-imdb-rating').text.replace('\n','')
    movie_name.append(moviename)
    movie_year.append(movieyear)
    rating.append(movierating)
top_100_movies_df = pd.DataFrame({'Movie Name': movie_name, 'Year Of Release': movie_year, 'Rating': rating})
top_100_movies_df.to_csv('question_2.csv')
top_100_movies_df
```

# Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
    a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
    a) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.

```python
urls = ['https://www.icc-cricket.com/rankings/womens/team-rankings/odi','https://www.icc-cricket.com/rankings/mens/team-rankings/odi']
ranking_type = []
team_name = []
total_matches = []
total_points = []
ratings = []
df = pd.DataFrame()
for i in urls:
    odi_response = requests.get(i)
    soup = BeautifulSoup(odi_response.content)
    rankingType = soup.select_one(".rankings-block__title-container > h4").text.strip('\\')
    team_name.append(soup.find('span', class_ = 'u-hide-phablet').text)
    total_matches.append(soup.find('td', class_='rankings-block__banner--matches').text)
    total_points.append(soup.find('td',class_ = 'rankings-block__banner--points').text)
    ratings.append(soup.find('td', class_ = 'rankings-block__banner--rating u-text-right').text.replace('\n',''))
    ranking_type.append(rankingType)
    tr_body = soup.find_all('tr', class_ = 'table-body')[0:9]
    for i in tr_body:
        teamName = i.find('span', class_ = 'u-hide-phablet').text
        team_name.append(teamName)
        ranking_type.append(rankingType)
    demo = soup.find_all("td", class_="table-body__cell u-center-text")[0:18]
    info = []
    for dta in demo:
        info.append(dta.text)
    for dta in range(0, len(info), 2):
        total_matches.append(info[dta])
    for dta in range(1, len(info), 2):
        total_points.append(info[dta])
    rating_1 = soup.find_all("td", class_="table-body__cell u-text-right rating")[0:9]
    for rt in rating_1:
        ratings.append(rt.text)

df['Ranking Type'] = ranking_type
df["Team Name"] = team_name
df["Total Matches"] = total_matches
df["Total Points"] = total_points
df["Ratings"] = ratings
df.to_csv('question_5_6_a_a.csv')
df
```

# Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
    b) Top 10 ODI Batsmen along with the records of their team and rating.
    c) Top 10 ODI bowlers along with the records of their team and rating.
    b) Top 10 women’s ODI Batting players along with the records of their team and rating.
    c) Top 10 women’s ODI all-rounder along with the records of their team and rating.

```python
urls = ['https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting', 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling','https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting','https://www.icc-cricket.com/rankings/womens/player-rankings/odi/bowling']
df = pd.DataFrame()
ranking_type = []
player_name = []
player_team = []
player_rating = []
for i in urls:
    odi_response = requests.get(i)
    soup = BeautifulSoup(odi_response.content)
    rankingType = soup.select_one(".rankings-block__title-container > h4").text.strip('\\')
    playerName = soup.find("td",class_="rankings-block__top-player-container")
    player_name.append(playerName.text.replace('\n',''))
    playerTeam = soup.find("div",class_="rankings-block__banner--nationality")
    player_team.append(playerTeam.text.replace('\n',''))
    playerRating = soup.find("td",class_="u-text-left")
    player_rating.append(playerRating.text.replace('\n',''))
    ranking_type.append(rankingType)
    info = []
    demo = soup.find_all("tr",class_="table-body")[0:9]
    for i in demo:
        playerName = i.find("td", class_="table-body__cell rankings-table__name name")
        player_name.append(playerName.a.text.replace('\n', ''))
        playerTeam = i.find("span", class_="table-body__logo-text")
        player_team.append(playerTeam.text.replace('\n', ''))
        playerRating = i.find("td", class_="table-body__cell rating")
        player_rating.append(playerRating.text.replace('\n', ''))
        ranking_type.append(rankingType)
df['Ranking Type'] = ranking_type
df["Player Name"] = player_name
df["Player Team"] = player_team
df["Player Rating"] = player_rating
df.to_csv('question_5_6_b_b_c_c.csv')
df
```

# Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.

```python
title = []
published = []
Content = []
video_link = []
coreyms_response = requests.get('https://coreyms.com/')
soup = BeautifulSoup(coreyms_response.content)
all_data = soup.find_all('article')
for data in all_data:
    title.append(data.h2.a.text)
    published.append(data.find('time', class_ = 'entry-time').text)
    Content.append(data.find('div', class_ = 'entry-content').p.text)
    try:
        data_video_link = data.find('iframe', class_ = 'youtube-player')['src']
        vid_lnk = f"https://youtube.com/watch?v={data_video_link.split('/')[4].split('?')[0]}"
    except Exception as ex:
        vid_lnk = 'No Link Found In Web-Site'
    video_link.append(vid_lnk)
coreyms_df = pd.DataFrame({'Title':title,'Published_Date':published,'Content':Content, 'Video Link':video_link})
coreyms_df.to_csv('question_7.csv')
coreyms_df
```

# Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
page = requests.get('https://www.dineout.co.in/bangalore-restaurants/dineout-pay')
soup = BeautifulSoup(page.content)
names = []
locations = []
cuisines = []
images = []
ratings = []
card_rest = soup.find_all('div', class_ = 'restnt-card restaurant')
for hotels in card_rest:
    names.append(hotels.find('div', class_="restnt-info cursor").text)
    locations.append(hotels.find('div', class_="restnt-loc ellipsis").text)
    cuisines.append(hotels.find('span', class_="double-line-ellipsis").text.split('| ')[1])
    images.append(hotels.find("img", class_="no-img")['data-src'])
    ratings.append(hotels.find('div', class_ = ['restnt-rating rating-2','restnt-rating rating-3','restnt-rating rating-4','restnt-rating rating-5']).text)

df = pd.DataFrame({'Restaurent Names':names,'Locations':locations,'Cuisines':cuisines, 'Rating':ratings, 'Images_url':images})
df.to_csv('question_9.csv')
df
```

# Write a python program to scrape first 10 product details which include product name , price , Image URL from
https://www.bewakoof.com/women-tshirts?ga_q=tshirts .
